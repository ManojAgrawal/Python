```python
import numpy as np
import pandas as pd
from datetime import datetime, date, time
```

Use data for user id 1

```python
df = df_orig.loc[df_orig.user_id==1]
df['Amount'] = abs(df['Amount'])
```

```python
df.head()
```

Create a new column with category and sub-category combination

```python
df['categ_subcat'] = df['category'] + "-" + df['sub_category']
```

Extract year, month and day from transaction date

```python
year = []
month = []
day = []
year_month = []

for row in df['transaction_date']:
    dt = datetime.strptime(row, "%Y-%m-%d %H:%M:%S")
    year.append(dt.year)
    month.append(dt.month)
    day.append(dt.day)
    year_month.append(str(dt.year) + str(dt.month) )
    
df['year'] = year
df['month'] = month
df['day'] = day  
df['year_month'] = year_month
```

```python
df.head()
```

Create tables for transaction amounts summarized by year-month, category and sub categories

```python
df_categ_yrmon = pd.DataFrame(df[df.transaction_type=="Debit"].groupby(['year_month','category']).Amount.sum()).reset_index()
df_subcateg_yrmon = pd.DataFrame(df[df.transaction_type=="Debit"].groupby(['year_month','categ_subcat']).Amount.sum()).reset_index()
```

```python
df_subcateg_yrmon.head(10)
```

Create a set of unique categories and find minimum and maximum amount limits

```python
subcategories = set(df_subcateg_yrmon['categ_subcat'])
subcategories_ls = []
sub_min_ls = []
sub_max_ls = []
for subcategory in subcategories:
    q75, q25 = np.percentile(df_subcateg_yrmon[df_subcateg_yrmon.categ_subcat==subcategory].Amount.dropna(), [75 ,25])
    iqr = q75 - q25
    min_val = q25 - (iqr*1.5)
    max_val = q75 + (iqr*1.5)
    subcategories_ls.append(subcategory)
    sub_min_ls.append(min_val)
    sub_max_ls.append(max_val)   
    
df_subcateg_limits = pd.DataFrame(
    {'categ_subcat': subcategories_ls,
     'min_limit': sub_min_ls,
     'max_limit': sub_max_ls
    })

df_subcateg_limits    
```

Create current year_month value for the user

```python
curr_dt = date.today()
curr_yrmon = str(curr_dt.year) + str(curr_dt.month)
curr_yrmon
```

Create spending tables for the user for current year/month by categories and subcategories

```python
df_subcateg_curr = df_subcateg_yrmon[df_subcateg_yrmon.year_month == curr_yrmon].copy()
df_subcateg_curr
```

Compare user's current monthly spending with max limits by sub-category and create sub-category alert dictionary

```python
subcateg_alert = dict()

for index,row in df_subcateg_curr.iterrows():
    for index2, row2 in df_subcateg_limits.iterrows():
        if row2.categ_subcat == row.categ_subcat:
            if row.Amount > row2.max_limit:
                subcateg_alert[row.categ_subcat] = row.Amount   
```

```python
categs = []
sub_cats = []
amounts = []

for key, value in subcateg_alert.items():
    categ, subcat = key.split("-")
    categs.append(categ)
    sub_cats.append(subcat)
    amounts.append(value)
    
alert_table    = pd.DataFrame(
    {'category': categs,
     'sub_category': sub_cats,
     'amount': amounts
    }) 
```

Create dataframe for average spend in each category for last 12 months

```python
categ_avg = pd.DataFrame(df_categ_yrmon.groupby(['category']).Amount.mean()).reset_index()
categ_avg
```

Create dataframe for average spend in each sub-category for last 12 months

```python
subcateg_avg_dict = df_subcateg_yrmon.groupby(['categ_subcat']).Amount.mean().to_dict()

categs = []
sub_cats = []
amounts = []

for key, value in subcateg_avg_dict.items():
    categ, subcat = key.split("-")
    categs.append(categ)
    sub_cats.append(subcat)
    amounts.append(value)
    
subcat_avg    = pd.DataFrame(
    {'category': categs,
     'sub_category': sub_cats,
     'avg amount': amounts
    })

subcat_avg
```

Create dataframe for user spending in the current year month period by subcategory

```python
df_curr_subcat = df_subcateg_yrmon[df_subcateg_yrmon.year_month == curr_yrmon]

categs = []
sub_cats = []
amounts = []

for index, row in df_curr_subcat.iterrows():
    categ, subcat = row.categ_subcat.split("-")
    categs.append(categ)
    sub_cats.append(subcat)
    amounts.append(row.Amount)

subcat_curr_amts    = pd.DataFrame(
    {'category': categs,
     'sub_category': sub_cats,
     'amount': amounts
    })

subcat_curr_amts   
```

Create dataframe for user spending in the current year month period by category

```python
cat_curr_amts  = pd.DataFrame(subcat_curr_amts.groupby(['category']).amount.sum()).reset_index()
cat_curr_amts
```
